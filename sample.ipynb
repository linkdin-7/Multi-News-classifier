```python
df['headline_description'] = df['short_description']
corpus=[]
invalid_data=[]
headline_description_corpus=[]
for i in range(0,len(df['headline_description'].iloc[:20000])):
    try:
        headline_description_corpus = re.sub('[^a-zA-Z]',' ',str(df['headline_description'][i]))
    except Exception as e:
        invalid_data.append(i)
        print (e,i)
        continue
    headline_description_corpus = headline_description_corpus.lower()
    headline_description_corpus = headline_description_corpus.split()
#     ps = nltk.PorterStemmer()
    headline_description_corpus = [ps.stem(word) for word in headline_description_corpus if not word in set(stopwords.words('english'))]
    headline_description_corpus = ' '.join(headline_description_corpus)
    corpus.append(headline_description_corpus)
```

```python
import time 
new_df = []
st = time.time()
for i in range(len(corpus)):
#     s = list(map(lambda x: " ".join([ps.stem(word) for word in  corpus if not word in set(stopwords.words('english'))]),[corpus[i]]))
    s = list(map(lambda x: " ".join([wn.lemmatize(word) for word in corpus[i].split(" ")]), [corpus[i]]))[0]
    new_df.append(s)
et = time.time()
print("Total time taken :", (et-st))
print(new_df)
```
